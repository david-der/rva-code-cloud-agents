```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read and process the data
df = pd.read_csv('./data/support_tickets_predicted_categories.csv')

# Convert created_at to datetime
df['created_at'] = pd.to_datetime(df['created_at'])

# Extract day of week and hour from created_at
df['day_of_week'] = df['created_at'].dt.day_name()
df['hour'] = df['created_at'].dt.hour

# Filter for business hours (8 AM to 6 PM)
df_business = df[df['hour'].between(8, 18)]

# Create pivot table for business hours
pivot_business = pd.crosstab(df_business['day_of_week'], df_business['hour'])

# Order days of week correctly
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
pivot_business = pivot_business.reindex(day_order)

# Set up the plotting
plt.figure(figsize=(12, 8))
plt.style.use('default')  # Use default style instead of dark background

# Create custom colormap from white to blue with transparency
colors = ['#FFFFFF00', '#E6F3FF', '#BFE1FF', '#99CFFF', '#66B2FF', '#3394FF', '#0066CC']
n_bins = 100
from matplotlib.colors import LinearSegmentedColormap
custom_blue = LinearSegmentedColormap.from_list('custom_blue', colors, N=n_bins)

# Create heatmap
sns.heatmap(pivot_business, 
            cmap=custom_blue,
            annot=True, 
            fmt='d',
            cbar_kws={'label': 'Number of Tickets'},
            annot_kws={'color': 'black', 'weight': 'bold'},
            )

# Customize text elements
plt.title('Support Ticket Distribution (Business Hours)', 
          pad=20, 
          color='black', 
          weight='bold')
plt.xlabel('Hour of Day', 
          color='black', 
          weight='bold')
plt.ylabel('Day of Week', 
          color='black', 
          weight='bold')

# Make tick labels black and bold
plt.xticks(color='black', weight='bold')
plt.yticks(color='black', weight='bold')

# Make colorbar label black and bold
cbar = plt.gcf().axes[-1]
cbar.set_ylabel('Number of Tickets', 
                color='black', 
                weight='bold')
cbar.tick_params(labelcolor='black')
for tick_label in cbar.get_yticklabels():
    tick_label.set_weight('bold')

# Adjust layout and save with white background
plt.tight_layout()
plt.savefig('support_tickets_business_hours.png', 
            dpi=300, 
            bbox_inches='tight', 
            transparent=True)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read and process the data
df = pd.read_csv('./data/support_tickets_predicted_categories.csv')

# Convert created_at to datetime
df['created_at'] = pd.to_datetime(df['created_at'])
df['hour'] = df['created_at'].dt.hour

# Filter for business hours (8 AM to 6 PM)
df_business = df[df['hour'].between(8, 18)]

# Create pivot table for categories by hour
pivot_categories = pd.crosstab(df_business['category'], df_business['hour'])

# Sort categories by total volume
category_totals = pivot_categories.sum(axis=1)
pivot_categories = pivot_categories.reindex(category_totals.sort_values(ascending=False).index)

# Set up the plotting
plt.figure(figsize=(12, 6))
plt.style.use('default')  # Use default style

# Create custom colormap from white to blue with transparency
colors = ['#FFFFFF00', '#E6F3FF', '#BFE1FF', '#99CFFF', '#66B2FF', '#3394FF', '#0066CC']
n_bins = 100
from matplotlib.colors import LinearSegmentedColormap
custom_blue = LinearSegmentedColormap.from_list('custom_blue', colors, N=n_bins)

# Create heatmap
sns.heatmap(pivot_categories, 
            cmap=custom_blue,
            annot=True, 
            fmt='d',
            cbar_kws={'label': 'Number of Tickets'},
            annot_kws={'color': 'black', 'weight': 'bold'})

# Customize text elements
plt.title('Support Ticket Categories by Business Hour', 
          pad=20, 
          color='black', 
          weight='bold')
plt.xlabel('Hour of Day', 
          color='black', 
          weight='bold')
plt.ylabel('Category', 
          color='black', 
          weight='bold')

# Make tick labels black and bold
plt.xticks(color='black', weight='bold', rotation=0)
plt.yticks(color='black', weight='bold')

# Make colorbar label black and bold
cbar = plt.gcf().axes[-1]
cbar.set_ylabel('Number of Tickets', 
                color='black', 
                weight='bold')
cbar.tick_params(labelcolor='black')
for tick_label in cbar.get_yticklabels():
    tick_label.set_weight('bold')

# Adjust layout and save with transparent background
plt.tight_layout()
plt.savefig('support_tickets_categories_business_hours.png', 
            dpi=300, 
            bbox_inches='tight', 
            transparent=True)

# Print some analysis
print("\nBusiness hours (8 AM - 6 PM) statistics:")
print("\nTotal tickets by category:")
print(category_totals.sort_values(ascending=False))

print("\nPeak business hours by category:")
for category in pivot_categories.index:
    peak_hour = pivot_categories.loc[category].idxmax()
    peak_count = pivot_categories.loc[category].max()
    print(f"{category}: Peak at {peak_hour}:00 with {peak_count} tickets")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
import seaborn as sns

# Read and process the data
df = pd.read_csv('./data/support_tickets_predicted_categories.csv')
df['created_at'] = pd.to_datetime(df['created_at'])
df['resolved_at'] = pd.to_datetime(df['resolved_at'])

# 1. Radar Chart: Category distribution by time blocks
# Create time blocks
df['time_block'] = pd.cut(df['created_at'].dt.hour,
                         bins=[0, 6, 12, 18, 24],
                         labels=['Night (0-6)', 'Morning (6-12)', 'Afternoon (12-18)', 'Evening (18-24)'])

# Calculate the values for each category and time block
radar_data = pd.crosstab(df['category'], df['time_block'])
radar_data = radar_data.div(radar_data.sum(axis=1), axis=0)  # Convert to percentages

# Set up the radar chart
fig = plt.figure(figsize=(15, 10))
gs = GridSpec(1, 2, width_ratios=[1.2, 0.8])

ax = fig.add_subplot(gs[0], projection='polar')

# Angles for each time block
angles = np.linspace(0, 2*np.pi, len(radar_data.columns), endpoint=False)
angles = np.concatenate((angles, [angles[0]]))  # Complete the circle

# Plot each category
for idx, category in enumerate(radar_data.index):
    values = radar_data.loc[category].values
    values = np.concatenate((values, [values[0]]))  # Complete the circle
    ax.plot(angles, values, 'o-', linewidth=2, label=category)
    ax.fill(angles, values, alpha=0.25)

# Customize the radar chart
ax.set_xticks(angles[:-1])
ax.set_xticklabels(radar_data.columns, color='black', weight='bold')
ax.set_title('Ticket Distribution by Category and Time Block', 
             pad=20, color='black', weight='bold')

# Add legend with correct parameters
legend = ax.legend(bbox_to_anchor=(1.5, 0.5), loc='center', 
                  frameon=True, edgecolor='black',
                  title='Categories')
plt.setp(legend.get_title(), fontsize=12, weight='bold')

# 2. Resolution Time Violin Plot
ax2 = fig.add_subplot(gs[1])

# Calculate resolution times in hours
df['resolution_time'] = (df['resolved_at'] - df['created_at']).dt.total_seconds() / 3600

# Create violin plot
sns.violinplot(data=df, y='category', x='resolution_time',
               ax=ax2, color='lightblue', orient='h')

# Customize violin plot
ax2.set_title('Resolution Time Distribution by Category',
              pad=20, color='black', weight='bold')
ax2.set_xlabel('Resolution Time (Hours)', color='black', weight='bold')
ax2.set_ylabel('Category', color='black', weight='bold')
ax2.tick_params(colors='black', labelcolor='black')
for label in ax2.get_xticklabels() + ax2.get_yticklabels():
    label.set_weight('bold')

# Set reasonable x-axis limits (e.g., up to 95th percentile)
x_limit = df['resolution_time'].quantile(0.95)
ax2.set_xlim(0, x_limit)

plt.tight_layout()
plt.savefig('support_tickets_advanced_viz.png', 
            dpi=300, bbox_inches='tight', transparent=True)

# Additional analysis: Print median resolution times
print("\nMedian Resolution Times by Category:")
median_times = df.groupby('category')['resolution_time'].median().sort_values(ascending=False)
for category, time in median_times.items():
    print(f"{category}: {time:.1f} hours")
```
